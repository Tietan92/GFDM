Import libraries

```python
import numpy as np                   # numpy module

from matplotlib import pyplot as plt # Pyplot module
plt.rc('font', size=5)               # Change General Font Size

# gfdm libraries
from linearHeatConduction2d import Preprocessor  # Linear Heat Conduction Pre-Processor
from linearHeatConduction2d import Postprocessor # Linear Heat Conduction Post-Processor
from gfdm2d import Mesh # Mesh Object
```

Define simulation- and material parameters

```python
frequency           = 40 # frequency of the simulation [1/100ms]
timeEnd             = 5 # End time of the simulation [*100ms]
numThreads          = 8 # num of threads used to run the simulation
specHeatCapacity    = 880 # specific heat capacity of the material [J/(kg K)]
thermalConductivity = 39 # thermal conductivity of the material [W/m K]
massDensity         = 3940 # mass density of the material [kg/mÂ³]
```

Preprocessing

```python
prepro  = Preprocessor("heating_element") # initialise preprocessor

prepro.setFrequency(frequency)
prepro.setTimeEnd(timeEnd)
prepro.setNumThreads(numThreads)
prepro.setSpecHeatCapacity(specHeatCapacity)
prepro.setThermalConductivity(thermalConductivity/10) # per 100ms
prepro.setMassDensity(massDensity)
```

Import and link mesh

```python
mesh = Mesh('mesh_31413.msh') # create mesh object
prepro.linkMesh(mesh)
fig, ax = plt.subplots()
fig.dpi = 200
fig.set_figheight(3)
fig.set_figwidth(3)
#mesh.plotMesh(ax, lw=0.3)
mesh.plotBoundary(ax, lw=2)
plt.legend()
```

Set boundary conditions

```python
prepro.setBoundaryCondition(physicalTag=39, boundaryType="Neumann", temp=0, filmCoeff=0, heatflux=0)
prepro.setBoundaryCondition(physicalTag=38, boundaryType="Dirichlet", temp=200)
```

Set initial conditions

```python
prepro.setInitialCondition(20)
```

Run solver

```python
solverPath = "YOUR SOLVER FOLDER/linearHeatConduction2d_BiCGSTAB"
prepro.setSolver(solverPath)
prepro.runSolver()
```

Postprocessing

```python

postpr = Postprocessor("heating_element") # Initialise postprocessor
postpr.linkMesh(mesh) # link mesh

# plot results
fig, ax = plt.subplots(nrows=1,ncols=3,figsize=(8,2), dpi=200)
postpr.plotResults(fig,ax,[0,1,5])
```

```python
#create animation of the simulation results
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(4,3), dpi=200)
postpr.triangulationPlotAnimated(fig,ax,"micro_heating_element_anim.mp4","jet")
```
