**Import libraries**

```python
import numpy as np                   # numpy module

from matplotlib import pyplot as plt # Pyplot module
plt.rc('font', size=5)               # Change General Font Size

# gfdm libraries
from incompressibleFlow2d import Preprocessor  # Pre-Processor
from incompressibleFlow2d import Postprocessor # Post-Processor
from gfdm2d import Mesh                        # Mesh Object
```

**Define simulation and material parameters**

```python
# Simulation parameters
frequency           = 30 # frequency of the simulation [1/s]
timeEnd             = 30 # End time of the simulation [s]
numThreads          = 6  # num of threads used to run the simulation (should not be higher than the maximum number of cpu cores)

# Material Parameters
# The incrompressible flow should have a Reynolds number of 1000 (Edge length of the square is 1m)
Re          = 1000  # Reynolds number       
massDensity = 1     # mass density [kg/mÂ³]
U           = 2     # inflow velocity [m/s]
dynViscosity = U/Re # dynamic viscosity is valculated in order to archieve the Reynolds number [kg/(ms)]
```

**Initialise Preprocessor**

```python
prepro  = Preprocessor("square_big_area") # Initialise preprocessor

prepro.setFrequency(frequency)       # set the simulation frequency
prepro.setTimeEnd(timeEnd)           # set the end-time of the simulation
prepro.setNumThreads(numThreads)     # set the number of threads
prepro.setMassDensity(massDensity)   # set up mass density
prepro.setDynViscosity(dynViscosity) # set up dynamic viscosity
#prepro.setSolverTolerance(1e-12)    # set up tolerance for iterative solver (if required)
```

**Import and link mesh**

```python
mesh = Mesh('square_10_25_5_10498.msh') # Mesh with 10498 nodes (The animation in Github is made with a mesh which has around 500.000 nodes)
prepro.linkMesh(mesh)                   # Link mesh to to the preprocessor object

# Plot boundaries
fig, ax = plt.subplots()
fig.dpi = 200
fig.set_figheight(2)
fig.set_figwidth(3)
#mesh.plotMesh(ax, lw=0.3)  # plot the mesh
mesh.plotBoundary(ax, lw=1) # plot the boundaries
plt.legend()
```

**Set boundary conditions**

```python
prepro.setBoundaryCondition(physicalTag=10, boundaryType='velocity', values=[U,0]) # Inflow boundary condition on the left side
prepro.setBoundaryCondition(physicalTag=11, boundaryType='outflow')                # Outflow boundary condition on the right side
prepro.setBoundaryCondition(physicalTag=12, boundaryType='slip')                   # Slip boundary condition upper side
prepro.setBoundaryCondition(physicalTag=13, boundaryType='slip')                   # Slip boundary condition down side
prepro.setBoundaryCondition(physicalTag=14, boundaryType='velocity', values=[0,0]) # No-slip boundary condition at the square
```

**Set initial condition**

```python
prepro.setInitialCondition(v0=[0,0]) # The velocity at the beginning should be 0
```

**Run solver**

```python
solverPath = "FOLDER WITH SOLVERS/incompressibleFlow2d_BiCGSTAB"
prepro.setSolver(solverPath) # Set the solver path
prepro.runSolver()           # Run the simulation
```

**Postprocessing**

```python
postpr = Postprocessor("square_big_area") # Initialise postprocessor
postpr.linkMesh(mesh) # link mesh
# plot results
fig, ax = plt.subplots(nrows=5,ncols=2,figsize=(10,15), dpi=100)
scale = np.linspace(0,1,10) 
postpr.plotResults(fig=fig,
                   ax=ax,
                   var='|vel|',          # Magnitude of the velocity should be plotted
                   tVal= scale*timeEnd,  # The velocity field at 10 different time-points between 0-tEnd should be plotted 
                   colormap='gist_gray') # see https://matplotlib.org/stable/users/explain/colors/colormaps.html
```

```python
#create animation of the simulation results
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(6,3), dpi=400)
postpr.triangulationPlotAnimated(fig = fig,
                                 ax = ax,
                                 var = '|vel|',               # Magnitude of the velocity should be animated
                                 name = "flow_along_square", # name of the .mp4 output file
                                 colormap = "gist_gray")          # see https://matplotlib.org/stable/users/explain/colors/colormaps.html
```
